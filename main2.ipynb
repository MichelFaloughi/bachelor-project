```python
from ParticleSystem import ParticleSystem
import pandas as pd
import os

WORLD_WIDTH = 100
WORLD_HEIGHT = 100
DELTA = 0.0001
MU = 0.4
EPSILON = 0.9
ALPHA = 0.8
DOT_SIZE = 3
MIDDLE_CLUSTER_SIZE = -1
NUM_ITERATIONS = 10 ** 18
INIT_REFRESH_RATE = 10
INIT_PAUSED_STATUS = False

file_name = f'M{MU}D{DELTA}E{EPSILON}A{ALPHA}_{WORLD_WIDTH}x{WORLD_HEIGHT}_first_cluster.xlsx'

# Ensure the Excel file has the correct column headers if it doesn't exist
if not os.path.isfile(file_name):
    df = pd.DataFrame(columns=["Run ID", "num iterations"])
    df.to_excel(file_name, index=False)  # Create the file with headers

for i in range(200):
    world = ParticleSystem(
        WORLD_WIDTH,
        WORLD_HEIGHT,
        DELTA,
        MU,
        EPSILON,
        ALPHA,
        DOT_SIZE,
        MIDDLE_CLUSTER_SIZE,
        NUM_ITERATIONS,
        INIT_REFRESH_RATE,
        INIT_PAUSED_STATUS
    )

    num_iterations_it_took = world.run_simulation_get_iteration_of_first_single_cluster(
        refresh_rate=280, check_rate=15000, init_render_status=False, first_check=50000
    )

    # Append directly to the file (avoiding reading the whole file)
    new_data = pd.DataFrame([[world.id, num_iterations_it_took]], columns=["Run ID", "num iterations"])

    with pd.ExcelWriter(file_name, mode='a', if_sheet_exists='overlay', engine='openpyxl') as writer:
        new_data.to_excel(writer, index=False, header=False, startrow=writer.sheets['Sheet1'].max_row)

    print(f'Finished World: {i}')
```

```python
# How many itereations before one cluster happends
from ParticleSystem import ParticleSystem
import pandas as pd
import os 

WORLD_WIDTH = 100 # 200 # 20
WORLD_HEIGHT = 100 # 200 # 20
DELTA = 0.0001 # 0.001 # 0.01 # 0.1 # probability to change direction
MU = 0.4 # density
EPSILON = 0.9 # probability to just follow the nomral direction for one iteration (and not randomly move)
ALPHA = 0.8 # probability to be an active (normal) particle
DOT_SIZE = 3 # 4
MIDDLE_CLUSTER_SIZE = -1 # 10
NUM_ITERATIONS =  10 ** 18 # None # 10000
INIT_REFRESH_RATE = 10 # 100
INIT_PAUSED_STATUS = False

file_name = f'M{MU}D{DELTA}E{EPSILON}A{ALPHA}_{WORLD_WIDTH}x{WORLD_HEIGHT}_first_cluster.xlsx'



for i in range(200):

    world = ParticleSystem(
        WORLD_WIDTH,
        WORLD_HEIGHT,
        DELTA,
        MU,
        EPSILON, 
        ALPHA,
        DOT_SIZE,
        MIDDLE_CLUSTER_SIZE,
        NUM_ITERATIONS,
        INIT_REFRESH_RATE,
        INIT_PAUSED_STATUS
    )

    num_iterations_it_took = world.run_simulation_get_iteration_of_first_single_cluster(

        refresh_rate=280, check_rate=60000, init_render_status=True, first_check=100000
        
        )
    
    if os.path.isfile(file_name): # the file already exists

        # read, write, push
        df = pd.read_excel(file_name)
        df.loc[len(df)] = [world.id, num_iterations_it_took] # adds to the df
        df.to_excel(file_name)
    
    else: # the file doesn't exist yet

        # create, write, push
        df = pd.DataFrame(columns=["Run ID", "num iterations"])
        df.loc[len(df)] = [world.id, num_iterations_it_took] # adds to the df
        df.to_excel(file_name)

    print(f'finished World: {i}')
```
