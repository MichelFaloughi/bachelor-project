```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define torus
theta = np.linspace(0, 2*np.pi, 100)  # Major radius angle
phi = np.linspace(0, 2*np.pi, 50)  # Minor radius angle
theta, phi = np.meshgrid(theta, phi)

R, r = 5, 1  # Major and minor radii
X = (R + r*np.cos(phi)) * np.cos(theta)
Y = (R + r*np.cos(phi)) * np.sin(theta)
Z = 0.2 * r * np.sin(phi)  # Flattened height

# Variable for ribbon width
ribbon_width = 2  # Adjust this value to make the ribbon wider/narrower

# Define a ribbon wrapping smoothly (with adjustable width)
t = np.linspace(0, 2*np.pi, 300)  # Full torus loop
w = np.linspace(-ribbon_width, ribbon_width, 30)  # Ribbon width controlled by variable

t, w = np.meshgrid(t, w)

# Helical ribbon path: Moves once around both minor and major radii
Xr = (R + r*np.cos(t)) * np.cos(t + np.pi/4) + w * np.cos(t)  
Yr = (R + r*np.cos(t)) * np.sin(t + np.pi/4) + w * np.sin(t)
Zr = 0.2 * r * np.sin(t)  

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

# Plot torus
ax.plot_surface(X, Y, Z, color='cyan', alpha=0.5, edgecolor='k')

# Plot wider ribbon with a bright red color
ax.plot_surface(Xr, Yr, Zr, color='red', alpha=1.0)  # Brighter red

# Adjust aspect ratio
ax.set_box_aspect([1,1,0.3])  # Reduce height in visualization

# Hide grid lines
ax.grid(False)

# Hide x, y, z axis
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

# Show the final plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define torus with a very reduced height
theta = np.linspace(0, 2*np.pi, 100)
phi = np.linspace(0, 2*np.pi, 50)
theta, phi = np.meshgrid(theta, phi)
R, r = 5, 1  # Reduce the minor radius r even more

X = (R + r*np.cos(phi)) * np.cos(theta)
Y = (R + r*np.cos(phi)) * np.sin(theta)
Z = 0.2 * r * np.sin(phi)  # Extra scaling factor to squash the height more

# Define a wrapped coil (helix around the torus)
t = np.linspace(0, 2*np.pi, 200)
Xr = (R + r*np.cos(5*t)) * np.cos(t)
Yr = (R + r*np.cos(5*t)) * np.sin(t)
Zr = 0.2 * r * np.sin(5*t)  # Also squashing the coil height

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

# Plot torus
ax.plot_surface(X, Y, Z, color='cyan', alpha=0.5, edgecolor='k')

# Plot ribbon/coil
ax.plot(Xr, Yr, Zr, color='red', linewidth=2)

# Adjust aspect ratio to make sure it looks truly flattened
ax.set_box_aspect([1,1,0.3])  # Reduce height in visualization



# Hide grid lines
ax.grid(False)

# Hide x, y, z axis
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
# ax.set_frame_on(False)  # Removes the box


plt.show()
```
